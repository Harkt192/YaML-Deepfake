```python

import os
import time
import warnings
import gc
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as T
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
from tqdm import tqdm

Image.MAX_IMAGE_PIXELS = None
warnings.filterwarnings('ignore')

os.environ['OMP_NUM_THREADS'] = '2'
os.environ['MKL_NUM_THREADS'] = '2'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb:128'
torch.backends.cudnn.benchmark = True
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True
torch.set_float32_matmul_precision('medium')

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"🖥 Устройство: {DEVICE}")
if DEVICE.type == 'cuda':
    print(f"   GPU: {torch.cuda.get_device_name(0)}")
    print(f"   Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")

class ChannelGate(nn.Module):
    def __init__(self, channels, reduction=8):
        super().__init__()
        self.gap = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, channels // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels, bias=False),
            nn.Sigmoid()
        )
    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.gap(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y

class SpatialGate(nn.Module):
    def __init__(self, kernel_size=5):
        super().__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size, padding=kernel_size//2, bias=False)
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        avg = torch.mean(x, dim=1, keepdim=True)
        maxv, _ = torch.max(x, dim=1, keepdim=True)
        concat = torch.cat([avg, maxv], dim=1)
        att = self.sigmoid(self.conv(concat))
        return x * att

class DynamicConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch, stride=1, use_gate=True):
        super().__init__()
        self.conv1 = nn.Conv2d(in_ch, out_ch, 3, stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_ch)
        self.conv2 = nn.Conv2d(out_ch, out_ch, 3, stride=1, padding=1, groups=out_ch, bias=False)
        self.bn2 = nn.BatchNorm2d(out_ch)
        self.conv3 = nn.Conv2d(out_ch, out_ch, 1, bias=False)
        self.bn3 = nn.BatchNorm2d(out_ch)
        self.relu = nn.ReLU(inplace=True)
        self.use_gate = use_gate
        if use_gate:
            self.ch_gate = ChannelGate(out_ch)
            self.sp_gate = SpatialGate()
        self.shortcut = nn.Sequential()
        if stride != 1 or in_ch != out_ch:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 1, stride, bias=False),
                nn.BatchNorm2d(out_ch)
            )
    def forward(self, x):
        residual = self.shortcut(x)
        out = self.conv1(x)
        out = self.bn1(out)
        out = self.relu(out)
        out = self.conv2(out)
        out = self.bn2(out)
        out = self.relu(out)
        out = self.conv3(out)
        out = self.bn3(out)
        if self.use_gate:
            out = self.ch_gate(out)
            out = self.sp_gate(out)
        out += residual
        out = self.relu(out)
        return out


class FrequencyBranch(nn.Module):
    def __init__(self, out_dim=128):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d(1)
        )
        self.fc = nn.Linear(64, out_dim)

    def forward(self, x):
        gray = x.mean(dim=1, keepdim=True)
        fft = torch.fft.fft2(gray, norm='ortho')
        amp = torch.abs(fft)
        amp = torch.log1p(amp)
        amp = (amp - amp.mean()) / (amp.std() + 1e-6)
        feat = self.conv(amp)
        feat = feat.flatten(1)
        del fft, amp, gray
        return self.fc(feat)

class FreqFusionNet(nn.Module):
    def __init__(self, num_classes=1):
        super().__init__()
        self.stem = nn.Sequential(
            nn.Conv2d(3, 32, 3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
        )
        self.block1 = DynamicConvBlock(32, 64, stride=2)
        self.block2 = DynamicConvBlock(64, 128, stride=2)
        self.block3 = DynamicConvBlock(128, 256, stride=2)
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        
        self.freq_branch = FrequencyBranch(out_dim=128)
        
        
        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(256 + 128, num_classes)
        
    def forward(self, x):
        original = x
        
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        spatial = self.global_pool(x).flatten(1)
        
        freq = self.freq_branch(original)
        
        combined = torch.cat([spatial, freq], dim=1)
        combined = self.dropout(combined)
        logits = self.fc(combined).squeeze(1)
        
        del original, x, spatial, freq, combined
        
        return logits

class FaceDataset(Dataset):
    def __init__(self, img_dir, filenames, labels, transform=None, preload=False):
        self.transform = transform
        img_dir = Path(img_dir)
        self.preload = preload
        self.paths = []
        
        for name in tqdm(filenames, desc="Индексация", leave=False):
            name = str(name)
            p = img_dir / (name + '.jpg')
            if not p.exists():
                p = img_dir / (name + '.png')
            if not p.exists():
                p = img_dir / name
            if p.exists():
                self.paths.append(p)
        
        self.labels = torch.tensor(labels, dtype=torch.float32)
        
        if preload and len(self.paths) <= 50000:
            self.images = []
            for path in tqdm(self.paths, desc="Предзагрузка", leave=False):
                img = Image.open(path).convert('RGB')
                self.images.append(img)
        else:
            self.images = None

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        if self.images is not None:
            img = self.images[idx].copy() if hasattr(self.images[idx], 'copy') else self.images[idx]
        else:
            img = Image.open(self.paths[idx]).convert('RGB')
        
        if self.transform:
            img = self.transform(img)
        
        return img, self.labels[idx]


train_transform = T.Compose([
    T.RandomResizedCrop(224, scale=(0.8, 1.0)),  
    T.RandomHorizontalFlip(p=0.5),
    T.RandomAffine(degrees=10, translate=(0.1, 0.1)),
    T.ColorJitter(brightness=0.1, contrast=0.1),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_transform = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])


class FocalLoss(nn.Module):
    def __init__(self, alpha=0.75, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
    def forward(self, inputs, targets):
        ce_loss = F.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal = (1 - pt) ** self.gamma * ce_loss
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        return (alpha_t * focal).mean()

def validate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    all_probs = []
    all_labels = []
    
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            with torch.cuda.amp.autocast():
                logits = model(imgs)
                loss = criterion(logits, labels)
                probs = torch.sigmoid(logits)
            
            total_loss += loss.item() * imgs.size(0)
            all_probs.append(probs.cpu())
            all_labels.append(labels.cpu())
            
            del imgs, labels, logits, loss, probs
    
    all_probs = torch.cat(all_probs)
    all_labels = torch.cat(all_labels)
    
    return total_loss / len(loader.dataset), all_probs, all_labels

def find_best_threshold(probs, labels):
    best_f1 = 0
    best_th = 0.5
    for th in np.linspace(0.3, 0.8, 20):
        preds = (probs >= th).float()
        tp = ((preds == 1) & (labels == 1)).sum().item()
        fp = ((preds == 1) & (labels == 0)).sum().item()
        fn = ((preds == 0) & (labels == 1)).sum().item()
        rec = tp / (tp + fn + 1e-8)
        prec = tp / (tp + fp + 1e-8)
        f1 = 2 * prec * rec / (prec + rec + 1e-8)
        if f1 > best_f1:
            best_f1 = f1
            best_th = th
    return best_th, best_f1

def prepare_loaders(data_dir, batch_size, val_split=0.15):
    csv_path = os.path.join(data_dir, 'train_solution.csv')
    df = pd.read_csv(csv_path)
    fnames = df.iloc[:, 0].astype(str).tolist()
    labels = df.iloc[:, 1].values.astype(np.float32)
    
    np.random.seed(42)
    indices = np.random.permutation(len(fnames))
    split_idx = int(len(indices) * (1 - val_split))
    train_idx = indices[:split_idx]
    val_idx = indices[split_idx:]
    
    train_ds = FaceDataset(
        os.path.join(data_dir, 'train_images'),
        [fnames[i] for i in train_idx],
        labels[train_idx],
        transform=train_transform,
        preload=False  
    )
    val_ds = FaceDataset(
        os.path.join(data_dir, 'train_images'),
        [fnames[i] for i in val_idx],
        labels[val_idx],
        transform=val_transform,
        preload=False
    )
    
    train_labels = labels[train_idx]
    class_counts = np.bincount(train_labels.astype(int))
    class_weights = 1.0 / (class_counts + 1e-8)
    sample_weights = class_weights[train_labels.astype(int)]
    sampler = WeightedRandomSampler(sample_weights, len(sample_weights))
    
    train_loader = DataLoader(
        train_ds, batch_size=batch_size, sampler=sampler,
        num_workers=0, pin_memory=True, persistent_workers=False
    )
    val_loader = DataLoader(
        val_ds, batch_size=batch_size, shuffle=False,
        num_workers=0, pin_memory=True, persistent_workers=False
    )
    
    return train_loader, val_loader


def train_model(data_dir='./dataset', epochs=120, batch_size=32, lr=1e-3):
    print("\n" + "="*70)
    print("🔥 FREQFUSIONNET — оптимизированная версия")
    print("="*70)
    
    train_loader, val_loader = prepare_loaders(data_dir, batch_size)
    print(f"✅ Train батчей: {len(train_loader)}, Val батчей: {len(val_loader)}")
    
    model = FreqFusionNet(num_classes=1)
    model = model.to(DEVICE)
    
    total_params = sum(p.numel() for p in model.parameters())
    print(f"📊 Модель: {total_params:,} параметров")
    
    df = pd.read_csv(os.path.join(data_dir, 'train_solution.csv'))
    labels_all = df.iloc[:, 1].values.astype(np.float32)
    pos_weight = (len(labels_all) - labels_all.sum()) / (labels_all.sum() + 1e-8)
    pos_weight = torch.tensor([pos_weight], dtype=torch.float32).to(DEVICE)
    
    criterion = FocalLoss(alpha=0.7, gamma=2.0)
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    
    total_steps = len(train_loader) * epochs
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer, max_lr=lr, total_steps=total_steps,
        pct_start=0.1, anneal_strategy='cos'
    )
    scaler = torch.cuda.amp.GradScaler() if DEVICE.type == 'cuda' else None
    
    history = {'train_loss': [], 'val_loss': [], 'val_f1': []}
    best_f1 = 0.0
    best_threshold = 0.5
    patience_counter = 0
    
    for epoch in range(epochs):
        start_time = time.time()
        
        model.train()
        total_loss = 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)
        
        for imgs, labels in pbar:
            imgs = imgs.to(DEVICE, non_blocking=True)
            labels = labels.to(DEVICE, non_blocking=True)
            
            optimizer.zero_grad(set_to_none=True)
            
            with torch.cuda.amp.autocast():
                logits = model(imgs)
                loss = criterion(logits, labels)
            
            if scaler:
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                scaler.step(optimizer)
                scaler.update()
            
            else:
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
            
            scheduler.step()
            total_loss += loss.item() * imgs.size(0)
            pbar.set_postfix({'loss': f'{loss.item():.4f}'})
            
            del imgs, labels, logits, loss
        
        train_loss = total_loss / len(train_loader.dataset)
        
        val_loss, val_probs, val_labels = validate(model, val_loader, criterion, DEVICE)
        best_th, val_f1 = find_best_threshold(val_probs, val_labels)
        
        preds = (val_probs >= best_th).float()
        tp = ((preds == 1) & (val_labels == 1)).sum().item()
        fp = ((preds == 1) & (val_labels == 0)).sum().item()
        fn = ((preds == 0) & (val_labels == 1)).sum().item()
        recall = tp / (tp + fn + 1e-8)
        precision = tp / (tp + fp + 1e-8)
        
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['val_f1'].append(val_f1)
        
        epoch_time = time.time() - start_time
        print(f"\n📊 Epoch {epoch+1}/{epochs} | Time: {epoch_time:.1f}s")
        print(f"   Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
        print(f"   Recall: {recall:.4f}, Precision: {precision:.4f}, F1: {val_f1:.4f}")
        
        if val_f1 > best_f1:
            best_f1 = val_f1
            best_threshold = best_th
            torch.save(model.state_dict(), 'best_freqfusionnet.pth')
            print(f"   💾 Новый лучший F1: {best_f1:.4f}")
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= 12:
                print(f"\n⏹️ Early stopping на эпохе {epoch+1}")
                break
        
        del val_probs, val_labels, preds
        if DEVICE.type == 'cuda':
            torch.cuda.empty_cache()
        gc.collect()
    
    # Графики
    print("\n📈 Построение графиков...")
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].plot(history['train_loss'], 'b-o', label='Train Loss')
    axes[0].plot(history['val_loss'], 'r-s', label='Val Loss')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()
    axes[0].grid(True)
    
    axes[1].plot(history['val_f1'], 'g-^', label='F1-Score')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('F1')
    axes[1].legend()
    axes[1].grid(True)
    plt.tight_layout()
    plt.show()
    
    print("\n" + "="*70)
    print(f"✅ ОБУЧЕНИЕ ЗАВЕРШЕНО!")
    print(f"🏆 Лучший F1: {best_f1:.4f} с порогом {best_threshold:.3f}")
    print("="*70)
    return best_threshold

if __name__ == '__main__':
    DATA_DIR = './dataset'
    
    if DEVICE.type == 'cuda':
        torch.cuda.empty_cache()
    gc.collect()
    
    best_th = train_model(data_dir=DATA_DIR, epochs=200, batch_size=32, lr=1e-3)
    
    print("\n🎉 Работа завершена!")
```
