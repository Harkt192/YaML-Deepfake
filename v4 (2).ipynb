```python
import os
import time
import warnings
import gc
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as T
import torchvision.transforms.functional as TF
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torch.nn.parallel import DataParallel
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
from tqdm import tqdm

Image.MAX_IMAGE_PIXELS = None
warnings.filterwarnings('ignore')

os.environ['OMP_NUM_THREADS'] = '2'
os.environ['MKL_NUM_THREADS'] = '2'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb:128'
torch.backends.cudnn.benchmark = True
torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
GPU_COUNT = torch.cuda.device_count() if torch.cuda.is_available() else 0
USE_MULTI_GPU = GPU_COUNT > 1

print(f"🖥️ Устройство: {DEVICE}")
if DEVICE.type == 'cuda':
    for i in range(GPU_COUNT):
        print(f"   GPU {i}: {torch.cuda.get_device_name(i)}")
    print(f"   Кол-во GPU: {GPU_COUNT}")


class DropBlock2D(nn.Module):
    """DropBlock - эффективная регуляризация для свёрточных слоёв"""
    def __init__(self, block_size=7, drop_prob=0.1):
        super().__init__()
        self.block_size = block_size
        self.drop_prob = drop_prob

    def forward(self, x):
        if not self.training or self.drop_prob == 0:
            return x
        gamma = self.drop_prob / (self.block_size ** 2)
        mask = torch.bernoulli(torch.ones_like(x[:, :1, :, :]) * gamma)
        mask = F.max_pool2d(mask, kernel_size=self.block_size, stride=1, 
                           padding=self.block_size//2)
        mask = 1 - mask
        x = x * mask
        x = x * (mask.numel() / mask.sum())
        return x

class StochasticDepth(nn.Module):
    """Stochastic Depth - случайное пропускание residual блоков"""
    def __init__(self, survival_prob=0.8):
        super().__init__()
        self.survival_prob = survival_prob

    def forward(self, x, identity):
        if not self.training:
            return x + identity
        if torch.rand(1) < self.survival_prob:
            return x + identity
        return identity

class SEBlock(nn.Module):
    """Squeeze-and-Excitation блок"""
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.gap = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, channels // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels, bias=False),
            nn.Sigmoid()
        )
    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.gap(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y

class EnhancedConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch, stride=1, use_dropblock=True, survival_prob=0.9):
        super().__init__()
        self.conv1 = nn.Conv2d(in_ch, out_ch, 3, stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_ch)
        self.conv2 = nn.Conv2d(out_ch, out_ch, 3, stride=1, padding=1, groups=out_ch, bias=False)
        self.bn2 = nn.BatchNorm2d(out_ch)
        self.conv3 = nn.Conv2d(out_ch, out_ch, 1, bias=False)
        self.bn3 = nn.BatchNorm2d(out_ch)
        self.se = SEBlock(out_ch)
        self.relu = nn.ReLU(inplace=True)
        
        self.use_dropblock = use_dropblock
        if use_dropblock:
            self.dropblock = DropBlock2D(block_size=5, drop_prob=0.1)
        
        self.stochastic_depth = StochasticDepth(survival_prob)
        
        self.shortcut = nn.Sequential()
        if stride != 1 or in_ch != out_ch:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 1, stride, bias=False),
                nn.BatchNorm2d(out_ch)
            )
    
    def forward(self, x):
        residual = self.shortcut(x)
        out = self.conv1(x)
        out = self.bn1(out)
        out = self.relu(out)
        out = self.conv2(out)
        out = self.bn2(out)
        out = self.relu(out)
        out = self.conv3(out)
        out = self.bn3(out)
        out = self.se(out)
        
        if self.use_dropblock:
            out = self.dropblock(out)
        
        out = self.stochastic_depth(out, residual)
        out = self.relu(out)
        return out

class FrequencyBranchImproved(nn.Module):
    def __init__(self, out_dim=128):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            DropBlock2D(block_size=3, drop_prob=0.05),
            nn.Conv2d(64, 128, 3, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d(1)
        )
        self.dropout = nn.Dropout(0.2)
        self.fc = nn.Linear(128, out_dim)

    def forward(self, x):
        gray = x.mean(dim=1, keepdim=True)
        fft = torch.fft.fft2(gray, norm='ortho')
        amp = torch.abs(fft)
        amp = torch.log1p(amp)
        amp = (amp - amp.mean()) / (amp.std() + 1e-6)
        feat = self.conv(amp)
        feat = feat.flatten(1)
        feat = self.dropout(feat)
        del fft, amp, gray
        return self.fc(feat)

class FreqFusionNetV2(nn.Module):
    def __init__(self, num_classes=1):
        super().__init__()
        self.stem = nn.Sequential(
            nn.Conv2d(3, 32, 3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            DropBlock2D(block_size=3, drop_prob=0.05),
        )
        
        # Основные блоки с возрастающей регуляризацией
        self.block1 = EnhancedConvBlock(32, 64, stride=2, survival_prob=0.95)
        self.block2 = EnhancedConvBlock(64, 128, stride=2, survival_prob=0.90)
        self.block3 = EnhancedConvBlock(128, 256, stride=2, survival_prob=0.85)
        self.block4 = EnhancedConvBlock(256, 512, stride=2, survival_prob=0.80)
        
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.freq_branch = FrequencyBranchImproved(out_dim=128)
        
        self.dropout1 = nn.Dropout(0.4)
        self.dropout2 = nn.Dropout(0.3)
        self.fc1 = nn.Linear(512 + 128, 256)
        self.fc2 = nn.Linear(256, num_classes)
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, x):
        original = x
        
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.block4(x)
        spatial = self.global_pool(x).flatten(1)
        
        freq = self.freq_branch(original)
        
        combined = torch.cat([spatial, freq], dim=1)
        combined = self.dropout1(combined)
        combined = self.relu(self.fc1(combined))
        combined = self.dropout2(combined)
        logits = self.fc2(combined).squeeze(1)
        
        del original, x, spatial, freq, combined
        return logits
    
def mixup_data(x, y, alpha=0.2):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    mixed_x = lam * x + (1 - lam) * x[index]
    y_a, y_b = y, y[index]
    return mixed_x, y_a, y_b, lam

def cutmix_data(x, y, alpha=1.0):
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1
    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)
    
    W, H = x.size(2), x.size(3)
    cut_rat = np.sqrt(1. - lam)
    cut_w = int(W * cut_rat)
    cut_h = int(H * cut_rat)
    cx = np.random.randint(W)
    cy = np.random.randint(H)
    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)
    
    x[:, :, bbx1:bbx2, bby1:bby2] = x[index, :, bbx1:bbx2, bby1:bby2]
    lam = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (W * H))
    y_a, y_b = y, y[index]
    return x, y_a, y_b, lam

# Усиленные аугментации
train_transform = T.Compose([
    T.RandomResizedCrop(224, scale=(0.7, 1.0)),
    T.RandomHorizontalFlip(p=0.5),
    T.RandomAffine(degrees=15, translate=(0.15, 0.15), scale=(0.85, 1.15)),
    T.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.05),
    T.RandomApply([T.GaussianBlur(kernel_size=3, sigma=(0.5, 1.5))], p=0.2),
    T.RandomApply([T.RandomAdjustSharpness(sharpness_factor=2)], p=0.15),
    T.ToTensor(),
    T.RandomErasing(p=0.25, scale=(0.02, 0.15), ratio=(0.3, 3.3)),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_transform = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])


class LabelSmoothingFocalLoss(nn.Module):
    """Комбинация Label Smoothing и Focal Loss"""
    def __init__(self, alpha=0.75, gamma=2.0, smoothing=0.05):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.smoothing = smoothing
    
    def forward(self, inputs, targets):
        targets = targets * (1 - self.smoothing) + 0.5 * self.smoothing
        
        ce_loss = F.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal = (1 - pt) ** self.gamma * ce_loss
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        return (alpha_t * focal).mean()

class RDropLoss(nn.Module):
    """R-Drop регуляризация - согласованность предсказаний"""
    def __init__(self, alpha=0.5):
        super().__init__()
        self.alpha = alpha
    
    def forward(self, logits1, logits2):
        p1 = F.log_softmax(logits1.unsqueeze(1), dim=-1)
        p2 = F.softmax(logits2.unsqueeze(1), dim=-1)
        kl_loss = F.kl_div(p1, p2, reduction='batchmean')
        return self.alpha * kl_loss

class FaceDataset(Dataset):
    def __init__(self, img_dir, filenames, labels, transform=None):
        self.transform = transform
        img_dir = Path(img_dir)
        self.paths = []
        
        for name in tqdm(filenames, desc="Индексация", leave=False):
            name = str(name)
            for ext in ['.jpg', '.png', '']:
                p = img_dir / (name + ext if ext else name)
                if p.exists():
                    self.paths.append(p)
                    break
        
        self.labels = torch.tensor(labels, dtype=torch.float32)

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        img = Image.open(self.paths[idx]).convert('RGB')
        if self.transform:
            img = self.transform(img)
        return img, self.labels[idx]

def train_one_epoch(model, loader, optimizer, criterion, scaler, epoch, 
                    use_mixup=True, use_cutmix=True, rdrop_loss=None):
    model.train()
    total_loss = 0
    pbar = tqdm(loader, desc=f"Epoch {epoch+1}", leave=False)
    
    for imgs, labels in pbar:
        imgs = imgs.to(DEVICE, non_blocking=True)
        labels = labels.to(DEVICE, non_blocking=True)
        
        aug_type = np.random.choice(['none', 'mixup', 'cutmix'], 
                                   p=[0.4, 0.35, 0.25])
        
        optimizer.zero_grad(set_to_none=True)
        
        with torch.cuda.amp.autocast():
            if aug_type == 'mixup' and use_mixup and epoch > 5:
                imgs, labels_a, labels_b, lam = mixup_data(imgs, labels)
                logits = model(imgs)
                loss = lam * criterion(logits, labels_a) + (1 - lam) * criterion(logits, labels_b)
            elif aug_type == 'cutmix' and use_cutmix and epoch > 5:
                imgs, labels_a, labels_b, lam = cutmix_data(imgs, labels)
                logits = model(imgs)
                loss = lam * criterion(logits, labels_a) + (1 - lam) * criterion(logits, labels_b)
            else:
                logits = model(imgs)
                loss = criterion(logits, labels)
            
            # R-Drop (двойной forward)
            if rdrop_loss is not None and epoch > 10:
                logits2 = model(imgs)
                loss += rdrop_loss(logits, logits2)
        
        if scaler:
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()
        else:
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
        
        total_loss += loss.item() * imgs.size(0)
        pbar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        del imgs, labels, logits, loss
    
    return total_loss / len(loader.dataset)

def validate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    all_probs = []
    all_labels = []
    
    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Val", leave=False):
            imgs = imgs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            with torch.cuda.amp.autocast():
                logits = model(imgs)
                loss = criterion(logits, labels)
                probs = torch.sigmoid(logits)
            
            total_loss += loss.item() * imgs.size(0)
            all_probs.append(probs.cpu())
            all_labels.append(labels.cpu())
            
            del imgs, labels, logits, loss, probs
    
    all_probs = torch.cat(all_probs)
    all_labels = torch.cat(all_labels)
    
    return total_loss / len(loader.dataset), all_probs, all_labels

def find_best_threshold(probs, labels):
    best_f1 = 0
    best_th = 0.5
    for th in np.linspace(0.2, 0.85, 30):
        preds = (probs >= th).float()
        tp = ((preds == 1) & (labels == 1)).sum().item()
        fp = ((preds == 1) & (labels == 0)).sum().item()
        fn = ((preds == 0) & (labels == 1)).sum().item()
        rec = tp / (tp + fn + 1e-8)
        prec = tp / (tp + fp + 1e-8)
        f1 = 2 * prec * rec / (prec + rec + 1e-8)
        if f1 > best_f1:
            best_f1 = f1
            best_th = th
    return best_th, best_f1

def prepare_loaders(data_dir, batch_size, val_split=0.15):
    csv_path = os.path.join(data_dir, 'train_solution.csv')
    df = pd.read_csv(csv_path)
    fnames = df.iloc[:, 0].astype(str).tolist()
    labels = df.iloc[:, 1].values.astype(np.float32)
    
    np.random.seed(42)
    indices = np.random.permutation(len(fnames))
    split_idx = int(len(indices) * (1 - val_split))
    train_idx = indices[:split_idx]
    val_idx = indices[split_idx:]
    
    train_ds = FaceDataset(
        os.path.join(data_dir, 'train_images'),
        [fnames[i] for i in train_idx],
        labels[train_idx],
        transform=train_transform
    )
    val_ds = FaceDataset(
        os.path.join(data_dir, 'train_images'),
        [fnames[i] for i in val_idx],
        labels[val_idx],
        transform=val_transform
    )
    
    train_labels = labels[train_idx]
    class_counts = np.bincount(train_labels.astype(int))
    class_weights = 1.0 / (class_counts + 1e-8)
    sample_weights = class_weights[train_labels.astype(int)]
    sampler = WeightedRandomSampler(sample_weights, len(sample_weights))
    
    train_loader = DataLoader(
        train_ds, batch_size=batch_size, sampler=sampler,
        num_workers=0, pin_memory=True, persistent_workers=False
    )
    val_loader = DataLoader(
        val_ds, batch_size=batch_size, shuffle=False,
        num_workers=0, pin_memory=True, persistent_workers=False
    )
    
    return train_loader, val_loader

def train_model(data_dir='./dataset', epochs=100, batch_size=32, lr=1e-3):
    print("\n" + "="*70)
    print("🔥 FREQFUSIONNET V2 — Защита от переобучения")
    print("="*70)
    
    train_loader, val_loader = prepare_loaders(data_dir, batch_size)
    print(f"✅ Train батчей: {len(train_loader)}, Val батчей: {len(val_loader)}")
    
    model = FreqFusionNetV2(num_classes=1)
    
    if USE_MULTI_GPU:
        print(f"🚀 Использование {GPU_COUNT} GPU")
        model = DataParallel(model)
        lr = lr * GPU_COUNT
    
    model = model.to(DEVICE)
    
    total_params = sum(p.numel() for p in model.parameters())
    print(f"📊 Модель: {total_params:,} параметров")
    
    criterion = LabelSmoothingFocalLoss(alpha=0.7, gamma=2.0, smoothing=0.05)
    rdrop_loss = RDropLoss(alpha=0.3)
    
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=5e-4)
    
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(
        optimizer, T_0=10, T_mult=2, eta_min=1e-6
    )
    
    scaler = torch.cuda.amp.GradScaler() if DEVICE.type == 'cuda' else None
    
    history = {'train_loss': [], 'val_loss': [], 'val_f1': []}
    best_f1 = 0.0
    best_threshold = 0.5
    
    for epoch in range(epochs):
        start_time = time.time()
        
        train_loss = train_one_epoch(
            model, train_loader, optimizer, criterion, scaler, epoch,
            use_mixup=True, use_cutmix=True, 
            rdrop_loss=rdrop_loss if epoch > 10 else None
        )
        
        val_loss, val_probs, val_labels = validate(model, val_loader, criterion, DEVICE)
        best_th, val_f1 = find_best_threshold(val_probs, val_labels)
        
        preds = (val_probs >= best_th).float()
        tp = ((preds == 1) & (val_labels == 1)).sum().item()
        fp = ((preds == 1) & (val_labels == 0)).sum().item()
        fn = ((preds == 0) & (val_labels == 1)).sum().item()
        recall = tp / (tp + fn + 1e-8)
        precision = tp / (tp + fp + 1e-8)
        
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['val_f1'].append(val_f1)
        
        scheduler.step()
        
        epoch_time = time.time() - start_time
        print(f"\n📊 Epoch {epoch+1}/{epochs} | Time: {epoch_time:.1f}s | LR: {scheduler.get_last_lr()[0]:.2e}")
        print(f"   Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
        print(f"   Recall: {recall:.4f}, Precision: {precision:.4f}, F1: {val_f1:.4f}")
        
        if val_f1 > best_f1:
            best_f1 = val_f1
            best_threshold = best_th
            if USE_MULTI_GPU:
                torch.save(model.module.state_dict(), 'best_freqfusionnet_v2.pth')
            else:
                torch.save(model.state_dict(), 'best_freqfusionnet_v2.pth')
            print(f"   💾 Новый лучший F1: {best_f1:.4f}")

        
        # Очистка памяти
        del val_probs, val_labels, preds
        if DEVICE.type == 'cuda':
            torch.cuda.empty_cache()
        gc.collect()
    
    # Графики
    print("\n📈 Построение графиков...")
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    
    axes[0].plot(history['train_loss'], 'b-o', label='Train Loss')
    axes[0].plot(history['val_loss'], 'r-s', label='Val Loss')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].set_title('Loss Dynamics')
    axes[0].legend()
    axes[0].grid(True)
    
    axes[1].plot(history['val_f1'], 'g-^', label='F1-Score')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('F1')
    axes[1].set_title('F1 Score')
    axes[1].legend()
    axes[1].grid(True)
    
    overfit = np.array(history['val_loss']) - np.array(history['train_loss'])
    axes[2].plot(overfit, 'r-o', label='Val Loss - Train Loss')
    axes[2].axhline(y=0, color='k', linestyle='--', alpha=0.5)
    axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('Loss Gap')
    axes[2].set_title('Overfitting Indicator')
    axes[2].legend()
    axes[2].grid(True)
    
    plt.tight_layout()
    plt.show()
    
    print("\n" + "="*70)
    print(f"✅ ОБУЧЕНИЕ ЗАВЕРШЕНО!")
    print(f"🏆 Лучший F1: {best_f1:.4f} с порогом {best_threshold:.3f}")
    if USE_MULTI_GPU:
        print(f"🚀 Использовано {GPU_COUNT} GPU")
    print("="*70)
    return best_threshold

if __name__ == '__main__':
    DATA_DIR = './dataset'
    
    if DEVICE.type == 'cuda':
        torch.cuda.empty_cache()
    gc.collect()
    
    base_batch_size = 24 if GPU_COUNT >= 2 else 32
    
    best_th = train_model(
        data_dir=DATA_DIR, 
        epochs=250, 
        batch_size=base_batch_size, 
        lr=1e-3
    )
    
    print("\n🎉 Работа завершена!")
```
